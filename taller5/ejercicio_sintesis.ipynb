# Taller 5:  Filtros lineales y no lineales

```python
PATH = "./imagenes/"
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread(PATH + "noised_circuit.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(imagen, cmap='gray')
plt.title("Imagen con ruido")
plt.axis('off')
plt.show()
```

```python
print("Tamaño de la imagen:", imagen.shape)
print("Valor mínimo de la imagen:", np.min(imagen))
print("Valor máximo de la imagen:", np.max(imagen))
```

```python
from skimage.filters.rank import median, majority, maximum, minimum, modal
from scipy.ndimage import convolve
import numpy as np

# Aplicar filtro de mediana 3x3
image_median = median(imagen, footprint=np.ones((3,3)))

fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].imshow(imagen, cmap='gray')
ax[0].set_title('Imagen con ruido (original)')
ax[0].axis('off')

ax[1].imshow(image_median, cmap='gray')
ax[1].set_title('Imagen filtrada con mediana 3x3')
ax[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Probar con diferentes tamaños de kernel de mediana
image_median_5x5 = median(imagen, footprint=np.ones((5,5)))
image_median_7x7 = median(imagen, footprint=np.ones((7,7)))

fig, ax = plt.subplots(1, 3, figsize=(18, 6))

ax[0].imshow(image_median, cmap='gray')
ax[0].set_title('Mediana 3x3')
ax[0].axis('off')

ax[1].imshow(image_median_5x5, cmap='gray')
ax[1].set_title('Mediana 5x5')
ax[1].axis('off')

ax[2].imshow(image_median_7x7, cmap='gray')
ax[2].set_title('Mediana 7x7')
ax[2].axis('off')

plt.tight_layout()
plt.show()
```
