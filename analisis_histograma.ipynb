## Cargue las imágenes de trabajo.

```python
import cv2

imagen_CLEF = cv2.imread(PATH_IMAGES + "CLEF.png", cv2.IMREAD_GRAYSCALE)
imagen_MUSCLE= cv2.imread(PATH_IMAGES + "MUSCLE.png", cv2.IMREAD_GRAYSCALE)
```

## Visualice los histogramas.

```python
import matplotlib.pyplot as plt
import cv2
# Para la imagen CLEF
hist = cv2.calcHist([imagen_CLEF],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

```

```python
# Para la imagen MUSCLE
hist = cv2.calcHist([imagen_MUSCLE],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

**Explicación histogramas:** Los histogramas permiten visualizar cómo se distribuyen los niveles de intensidad de píxeles en una imagen. En la imagen CLEF, el histograma muestra una mayor de píxeles en los niveles de intensidad cercanos a los valores medios de la distribución, lo que indica que la imagen tiene una intensidad de tonos grises, no muy oscuros o muy claros. En contraste, el histograma de la imagen MUSCLE presenta una mayor variedad de tonos y un contraste más pronunciado en la imagen. Esto se puede notar en los diferentes picos que tiene la imagen.
