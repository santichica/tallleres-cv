## En la imagen muscle.png, se desea colorear de blanco (255) las fibras rojas (representadas por un gris oscuro). Tenga en cuenta que el resto de la imagen debe quedar inalterado. Es decir, las fibras rojas quedan blancas pero lo demás no se altera. Realice esta operación utilizando ÚNICAMENTE las operaciones vistas en esta parte. Para esta solución NO puede usar mapas de colores.

Iniciamos cargando la imagen y visualizandola.

```python
# Modificar las rutas de las imagenes.
PATH_IMAGES = "./imagenes/"
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

imagen = cv2.imread(PATH_IMAGES + "MUSCLE.png", cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(imagen, cmap='gray')
plt.show()
```

Ahora veremos el histograma de la imagen.

```python
hist = cv2.calcHist([imagen],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

La estrategia será utilizar una máscara que mapee los valores inferiores a un threshold a 255 (blanco) y el resto al mismo valor.

```python
# crear vector de mapeo donde los valores menores al threshold se asignan a 255 hazlo pydanticamente con list comprehension
def crear_vector(threshold):
    return np.array([255 if i < threshold else i for i in range(256)], dtype='uint8')
```

```python
umbral = crear_vector(100)
```

```python
imagen_transformada = cv2.LUT(imagen, umbral)
```

```python
plt.imshow(imagen_transformada, cmap='gray')
plt.show()
```

Cambiamos el umbral para acercarnos más a la imagen del ejercicio.

```python
umbral = crear_vector(90)
imagen_transformada = cv2.LUT(imagen, umbral)
plt.imshow(imagen_transformada, cmap='gray')
plt.show()
```

Vemos que el histograma cambia considerablemente y los valores de intensidad 255 cambian la escala del mismo.

```python
hist = cv2.calcHist([imagen_transformada],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

El resultado final logra imitar la imagen pedida en el ejercicio.
